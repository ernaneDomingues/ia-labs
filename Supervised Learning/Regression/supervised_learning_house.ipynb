```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harlfoxem/housesalesprediction")

print("Path to dataset files:", path)
```

```python
for file in os.listdir(path):
    if file.endswith(".csv"):
        csv_path = os.path.join(path, file)
        break
```

```python
df = pd.read_csv(csv_path)
```

```python
df_copy = df[['price','bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'yr_built', 'yr_renovated', 'zipcode', 'condition']]
```

```python
X = df_copy.drop(columns=['price'], axis=1)
y = df_copy['price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error
```

```python
model_scaler = StandardScaler()
X_scaled = model_scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
def metrics_report(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"MAPE: {mape}")
```

### Regression Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

```python
y_pred_lr = model_lr.predict(X_test)

metrics_report(y_test, y_pred_lr)
```

### KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model_knn = KNeighborsRegressor()
model_knn.fit(X_train, y_train)
```

```python
y_pred_knn = model_knn.predict(X_test)
metrics_report(y_test, y_pred_knn)
```

### Support Vector Regressor

```python
from sklearn.svm import SVR
```

```python
model_svr = SVR()
model_svr.fit(X_train, y_train)
```

```python
y_pred_svr = model_svr.predict(X_test)
metrics_report(y_test, y_pred_svr)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dt = DecisionTreeRegressor()
model_dt.fit(X_train, y_train)

```

```python
y_pred_dt = model_dt.predict(X_test)
metrics_report(y_test, y_pred_dt)
```

### XGBoost Regressor

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor()
model_xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = model_xgb.predict(X_test)
metrics_report(y_test, y_pred_xgb)
```
